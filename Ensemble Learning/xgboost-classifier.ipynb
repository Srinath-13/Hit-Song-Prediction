```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import cross_val_score, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score
```

## Import Data

```python
X=pd.read_csv("/kaggle/input/hit-prediction-processed-data/Hit Prediction/X_selected.csv")
y=np.load("/kaggle/input/hit-prediction-processed-data/Hit Prediction/y_selected.npy")
```

## Instantiate XGBoost

```python
xgb_model = xgb.XGBClassifier(objective='binary:logistic', random_state=42)
```

## Define Param Grid and Stratified CV

```python
param_grid = {
    'learning_rate': [0.1, 0.3, 0.5],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 150],
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

## Find Optimal Hyper Parameters

```python
grid_search = GridSearchCV(xgb_model, param_grid, cv=cv)
grid_search.fit(X,y)
```

## Best Fit

```python
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```
