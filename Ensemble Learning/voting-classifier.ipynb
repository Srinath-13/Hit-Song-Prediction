### Import Libraries

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
```

### Load Dataset

```python
X_selected=pd.read_csv("/kaggle/input/hit-prediction-processed-data/Hit Prediction/X_selected.csv")
y_selected=np.load("/kaggle/input/hit-prediction-processed-data/Hit Prediction/y_selected.npy")
```

### Base Models for Voting Classifier

Only KNN, SVM and Random Forest models are selected for the Voting Classifier since their performance is similar based on the ROC and AUC 

```python
knn = KNeighborsClassifier(algorithm='ball_tree',n_neighbors=19,p=1,weights='distance')

svm = SVC(C=1.5, class_weight='balanced', gamma='scale', kernel='rbf',probability=True)

rf = RandomForestClassifier(max_depth=20, max_features='log2', n_estimators=300, oob_score=True, ccp_alpha=0.0004)
```

### Hard Voting Classifier

#### Model Building

```python
eclf = VotingClassifier(estimators=[('knn', knn), ('rf', rf), ('svc', svm)], voting='hard')
```

#### Cross Validation - Hard Voting

```python
scores = cross_val_score(eclf, X_selected, y_selected, cv=5)
scores
```

```python
print("%0.5f accuracy with a standard deviation of %0.5f" % (scores.mean(), scores.std()))
```

### Soft Voting Classifier

#### Model Building

```python
eclf = VotingClassifier(estimators=[('knn', knn), ('rf', rf), ('svc', svm)], voting='soft')
```

#### Cross Validation - Soft Voting

```python
scores = cross_val_score(eclf, X_selected, y_selected, cv=5)
scores
```

```python
print("%0.5f accuracy with a standard deviation of %0.5f" % (scores.mean(), scores.std()))
```
